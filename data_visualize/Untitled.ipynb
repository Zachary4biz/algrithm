### 使用 seaborn作图

```python
# encoding=utf-8
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
#%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

- kde     是高斯分布密度图，绘图在0-1之间
- hist    是否画直方图
- rug     在X轴上画一些分布线
- fit     可以制定某个分布进行拟合
- label   legend时的值
- axlabel 制定横轴的说明


&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<font color=gray size=2>使用"plt.show()"，sns.plt.show()是BUG，seaboran作者已经在0.8版本修复</font>

```python
tips['total_bill']
```

```python
tips['tip']
```

```python
sns.distplot(tips['total_bill'],bins=None)
sns.set()
plt.show()
```

```python
sns.distplot(tips['total_bill'],bins=None,hist=False)
sns.set()
plt.show()
```

- 控制纵轴是否使用密度分布0到1

```python
# 这里如果hist=False没有图像；kde控制纵轴是否是密度分布（0到1之间）
sns.distplot(tips['total_bill'],bins=None,hist=True,kde=False) 
sns.set()
plt.show()
```

- 使用gamma分布来 “拟合分布“ 并去除了 密度分布

```python
from scipy import stats
sns.distplot(tips.total_bill,fit=stats.gamma,kde=False)
plt.show()
```

- 使用gamma分布，同时开启了密度分布，看到蓝色的线就是密度分布

```python
from scipy import stats
sns.distplot(tips.total_bill,fit=stats.gamma,kde=True)
plt.show()
```
